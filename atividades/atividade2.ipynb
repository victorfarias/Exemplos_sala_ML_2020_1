```python
import numpy as np
data = np.loadtxt("./housing.data")
np.random.shuffle(data)
X = data[:, -2]
y = data[:, -1]
```

```python
n = X.shape[0]
n_train = int(n*0.8)
n_test = n - n_train
X_train = X[:n_train]
X_test = X[-n_test:]
y_train = y[:n_train]
y_test = y[-n_test:]
```

```python
from atividade1 import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred_train = reg.predict(X_train)
y_pred_test = reg.predict(X_test)
```

```python
from atividade1 import MAE
# Usar R2 tb
print(MAE(y_pred_train, y_train))
print(MAE(y_pred_test, y_test))
```

```python
import matplotlib.pyplot as plt
x_reta = np.array([0., 40.])
y_reta = reg.predict(x_reta)


plt.scatter(X,y)
plt.plot(x_reta, y_reta)

plt.show()
```

```python
plt.scatter(y_pred_test, y_test)
plt.plot([0,50], [0,50])
plt.show()
```

```python
X_ = np.c_[X, X*X]
X_train_ = X_[:n_train]
X_test_ = X_[-n_test:]
```

```python
reg_ = LinearRegression()
reg_.fit(X_train_, y_train)
y_pred_train_ = reg_.predict(X_train_)
y_pred_test_ = reg_.predict(X_test_)

print(MAE(y_train, y_pred_train_))
print(MAE(y_test, y_pred_test_))
```

```python
x_reta_ = np.linspace(0., 45.)
x_reta_2 = np.c_[x_reta_ , x_reta_*x_reta_]
# print(x_reta_)
y_reta_ = reg_.predict(x_reta_2)

plt.scatter(X,y)
plt.plot(x_reta_, y_reta_)

plt.show()
```
