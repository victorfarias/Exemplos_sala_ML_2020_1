```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,4])
y = np.array([2,5])
plt.plot(x, y)
plt.show()
```

```python
def f(x):
    return 6*x*x + 2*x + 2

x = np.linspace(0., 10., num=50)
y = f(x)
plt.plot(x,y)
plt.show()
```
